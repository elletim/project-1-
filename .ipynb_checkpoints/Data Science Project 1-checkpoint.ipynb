```python
import pandas as pd
import requests as req
from io import StringIO
import matplotlib.pyplot as plt 


la = req.get("http://berkeleyearth.lbl.gov/air-quality/maps/cities/United_States/California/Los_Angeles.txt") 
la_df= pd.read_csv(StringIO(la.text),skiprows=10,sep='\t')
la_df.columns= ['Year','Month','Day', 'UTC Hour','PM2.5','PM10_mask','Retrospective']     

print(la_df.head())
```

```python
la1_df = la_df[["Year","Month","PM2.5"]]
la2_df =la1_df[(la_df["Year"].isin([2018,2019,2020,2021,2022])) & (la_df["Month"].isin([3,4,5]))]
la3_df = la2_df.groupby(["Year","Month"])
la4_df = la3_df["PM2.5"].mean()
la4_df

la4_df.plot()
```
